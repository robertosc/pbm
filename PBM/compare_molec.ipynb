```python
path_p = "/mnt/c/Users/rober/OneDrive/Documentos/UCREA/obabel-mol2s/"
files = glob.glob(path_p + "*.mol2", recursive=True)

#file_names = [os.path.basename(i) for i in files]
#print(file_names)

similar_compounds = {}
for i in range(len(files)):
    compounds = []
    name1 = os.path.basename(files[i])
    mol = rd.rdmolfiles.MolFromMol2File(files[i])
    mfp1 = rd.rdMolDescriptors.GetMorganFingerprint(mol, 2)
    for j in range(i+1, len(files)):
        name2 = os.path.basename(files[j])
        mol1 = rd.rdmolfiles.MolFromMol2File(files[j])
        mfp2 = rd.rdMolDescriptors.GetMorganFingerprint(mol1, 2)
        comp = rd.DataStructs.DiceSimilarity(mfp1,mfp2)
        if(comp > 0.6):
            compounds.append({re.sub(r"(_out)?.mol2", "", os.path.basename(files[j])):comp})
        #print(f"Morgan fingerprint comparison from {name1} vs {name2}: {rd.DataStructs.DiceSimilarity(mfp1,mfp2)}\n")
    if len(compounds) > 0:
        similar_compounds[re.sub(r"(_out)?.mol2", "", os.path.basename(files[i]))] = compounds

for i in similar_compounds:
    print(i, ":", similar_compounds[i], '\n')
```
