```python
def image_to_HSV(data_object):
    max_1=data_object.max()
    for i in range(data_object.shape[0]):
        for j in range(data_object.shape[1]):
            data_object[i, j] = data_object[i, j]/max_1
    plt.imshow(data_object)
            
image_to_HSV(tif_data_1)
```
